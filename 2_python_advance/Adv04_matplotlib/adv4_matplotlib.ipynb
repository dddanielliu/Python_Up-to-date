## [matplotlib](http://matplotlib.org)
資料視覺化非常強大的套件

```python
import numpy as np
import matplotlib.pyplot as plt
```

**一般直線圖**

```python
#例題 畫一條 y=1+2πx的圖 （x從0~2 step0.01）
x = np.arange(0.0, 2.0, 0.01)
y = 1 + 2 * np.pi * x
plt.plot(x,y)
plt.title("Line plot")
plt.show()
```

```python
#演練 畫一條 y=2x^2-6x+10的圖 （x從0~50 step0.01）
x = np.arange(0,50,0.01)
y = 2*x**2-6*x+10
plt.plot(x,y)
plt.show()
```

**散布圖-scatter plot**

```python
#例題 y=1+2πx （x從0~15 step1）

X = np.arange(0, 15)
Y = 1 + 2 * np.pi * X
plt.scatter(X, Y)
plt.title("Scatter plot")
plt.show()
```

```python
#演練 畫 y=2x^2-6x+10的散布圖 （x從0~15 step1）
x = np.arange(0,15)
y = 2*x**2-6*x+10
plt.scatter(x,y)
plt.show()
```

**直方圖-histogram**

```python
#例題
x = np.random.randn(10000)  #隨機產生常態分佈

##the histogram of the data
plt.hist(x, 10, density=True, facecolor='g')
plt.show()
```

```python
# 例題
x = np.random.randn(1000000)

# the histogram of the data
plt.hist(x, 100, density=True, facecolor='g')
plt.show()

```

## 額外小套件 datetime
專門處理時間格式的套件，也很常在pandas或是matplotlib裡面被應用

```python
import datetime
```

datetime.date.today():今天日期

datetime.datetime.now():現在時間

```python
print(datetime.date.today())
print(datetime.datetime.now())
```

```python
# date(year, month, day)
#例題
datetime.date(2019,11,11)
```

```python
#演練 用你的生日以datetime.date呈現
datetime.date(1994,5,22)
```

```python
# datetime(year, month, day[, hour[, minute[, second[, microsecond[,tzinfo]]]]])
#例題
print(datetime.datetime(2009,5,10,10,10,10))
```

```python
#演練 以datetime.datetime呈現今天早上8點50分
print(datetime.datetime(2019,11,18,8,50))
```

datetime.timedelta：計算時間

timedelta只能指定到"週"以下（週、天、小時、分鐘、秒、毫秒）

```python
datetime.datetime.now() + datetime.timedelta(hours=100)
```

```python
#例題3
t1 = datetime.date(year = 2018, month = 7, day = 12)
t2 = datetime.date(year = 2017, month = 12, day = 23)
t1 - t2
```

```python
#演練3 計算出你的生日到今天過了幾天
t1 = datetime.date(year = 1994, month = 5, day = 22)
t2 = datetime.date.today()
t2-t1
```

strftime()：datetime object to string

strptime()：string to datetime object

```python
print(datetime.datetime.now().strftime("%H:%M:%S"))
print(datetime.datetime.now().strftime("%m/%d/%Y, %H:%M:%S"))
```

```python
print(datetime.datetime.strptime("21 June, 2018", "%d %B, %Y"))
```
