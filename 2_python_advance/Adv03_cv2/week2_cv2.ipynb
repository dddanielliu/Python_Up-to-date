```python
#例題 灰階處理
img_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
print(img_gray.shape)
plt.imshow(img_gray,cmap='gray')
```

## 進階功能簡介 (optional)

```python
#邊緣偵測
edges = cv2.Canny(lenna,230,255)
plt.imshow(edges,cmap='gray')
```

```python
#增加畫質
res = cv2.resize(lenna,None,fx=2, fy=2, interpolation = cv2.INTER_CUBIC)
print(res.shape)
plt.imshow(res)
```

```python
#變形工具
rows,cols,ch = lenna.shape
pts1 = np.float32([[10,50],[200,50],[50,200]])
pts2 = np.float32([[100,100],[200,50],[100,250]])

M = cv2.getAffineTransform(pts1,pts2)

dst = cv2.warpAffine(lenna,M,(cols,rows))
plt.imshow(dst)
```
